```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Transmission'].value_counts()
    for x_label, grp in df.groupby('Fuel_Type')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Fuel_Type')
_ = plt.ylabel('Transmission')
```

**Question C**

```python
df = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission'])
```

```python
df
```

**Question D**

```python
from datetime import datetime
# Calculate current year
current_year = datetime.now().year

# Calculate the age of the car
df['Age'] = current_year - df['Year']
```

```python
df
```

```python
# Add 'Fit_Test' column
df['Fit_Test'] = df['Kilometers_Driven'].apply(lambda x: 'Fit' if x < 25000 else 'Unfit')
```

```python
df.head()
```

```python
# Count the occurrences of each category in 'Fit_Test' column
fit_test_counts = df['Fit_Test'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(fit_test_counts, labels=fit_test_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen'])
plt.title('Fit Test Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
selected_cols = df[['Name', 'Owner_Type', 'Mileage', 'Engine', 'Seats']]

print(selected_cols)
```

```python
re_df = df.rename(columns={'Unnamed: 0': 'ID', 'Kilometers_Driven': 'ODO'})

print(re_df)
```

```python
df['Price_in_lakhs'] = df['Price'] * 100000

print("\nMutated DataFrame")
print(df)
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
sns.pairplot(df[['Mileage', 'Engine', 'Power', 'Price_in_lakhs']])
plt.suptitle('Pairplot of Mileage, Engine, Power, and Price', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='Location', y='Price_in_lakhs', data=df)
plt.title('Price Distribution by Location')
plt.xlabel('Location')
plt.ylabel('Price (in Lakhs)')
plt.xticks(rotation=45)
plt.show()
```

```python
sorted_data = df.sort_values(by='Age', ascending=False)

print(sorted_data)
```

```python
# Grouping by Fuel Type and Calculating Mean Price:
fuel_type_price = df.groupby('Fuel_Type_Diesel')['Price_in_lakhs'].mean()
print(fuel_type_price)
```

```python
# Grouping by Transmission Type and Finding Maximum Power
transmission_power_max = df.groupby('Transmission_Manual')['Power'].max()
print(transmission_power_max)
```

```python
import matplotlib.pyplot as plt
grouped_cars_by_location = df.groupby('Location').size().reset_index(name='Number_of_Cars')

print(grouped_cars_by_location)
```

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Location', y='Number_of_Cars', data=grouped_cars_by_location)
plt.title('Number of Cars by Location')
plt.xlabel('Location')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()
```
